## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import datetime
```

```python
df=pd.read_csv('C:/Users/Documents/')
```

```python
#Melting original and creating column 'day'
table=pd.melt(df,['id','year','month','element'],var_name='day')
table.head(50)
```

```python
#Use pivot_table to add tmax and tmin to columns
table = table.pivot_table(index=["id","year","month","day"], columns="element", values="value")
table.reset_index(drop=False, inplace=True)
```
