```python
ifra_sum = datab_infra.sum(axis = 1)
ifra_sum
```

```python
ifra_sum_da = pd.DataFrame(ifra_sum, columns = ['total_of_Infrastructures'])
ifra_sum_da
```

```python
#cutoffs quantis

labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']

n = pd.qcut(ifra_sum_da['total_of_Infrastructures'],5, labels=labels)
```
