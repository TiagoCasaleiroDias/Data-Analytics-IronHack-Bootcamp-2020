```python
import pandas as pd 
import numpy as np
import geopy
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="x", timeout=None)
```

```python
mus = pd.read_csv('museus_original.csv')
mus
```

```python

mus['FREGUESIA'] = [(geolocator.reverse(str(mus['Y'][i]) + ', ' + str(mus['X'][i]))) for i in range(len(mus['Y']))]
```

```python
mus
```

```python

mus['FREGUESIA'] = [(str(mus['FREGUESIA'][p]).split(','))[z].lstrip() for p in range(len(mus['FREGUESIA'])) for z in range( len( str(mus['FREGUESIA'][p]).split(',')) -1 ) if (str(mus['FREGUESIA'][p]).split(','))[z+1] == ' Lisboa']
```

```python
mus = mus[['FREGUESIA','INF_SITE']]
```

```python
mus=(mus.groupby(['FREGUESIA'],as_index=False).count()).rename(columns={"INF_SITE": "f_Museums"})
```

```python
mus
```

```python
mus.to_csv('Museus_frequency.csv')
```
