```python
import pandas as pd 
import numpy as np
import geopy
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="x", timeout=None)
```

```python
col = pd.read_csv('College_original.csv')
col
```

```python

col['FREGUESIA'] = [(geolocator.reverse(str(col['Y'][i]) + ', ' + str(col['X'][i]))) for i in range(len(col['Y']))]
```

```python
col
```

```python

col['FREGUESIA'] = [(str(col['FREGUESIA'][p]).split(','))[z].lstrip() for p in range(len(col['FREGUESIA'])) for z in range( len( str(col['FREGUESIA'][p]).split(',')) -1 ) if (str(col['FREGUESIA'][p]).split(','))[z+1] == ' Lisboa']
```

```python
col = col[['FREGUESIA','INF_SITE']]
```

```python
col=(col.groupby(['FREGUESIA'],as_index=False).count()).rename(columns={"INF_SITE": "f_College"})
```
