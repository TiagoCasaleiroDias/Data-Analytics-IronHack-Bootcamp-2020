```python
import pandas as pd 
import numpy as np
import geopy
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="x", timeout=None)
```

```python
pol = pd.read_csv('policy_original.csv')
pol
```

```python
pol['FREGUESIA'] = [(geolocator.reverse(str(pol['Y'][i]) + ', ' + str(pol['X'][i]))) for i in range(len(pol['Y']))]

```

```python
pol
```

```python
pol['FREGUESIA'] = [(str(pol['FREGUESIA'][p]).split(','))[z].lstrip() for p in range(len(pol['FREGUESIA'])) for z in range( len( str(pol['FREGUESIA'][p]).split(',')) -1 ) if (str(pol['FREGUESIA'][p]).split(','))[z+1] == ' Lisboa']
```

```python
pol = pol[['FREGUESIA','COD_SIG']]
```

```python
pol=(pol.groupby(['FREGUESIA'],as_index=False).count()).rename(columns={"COD_SIG": "f_Police_Office"})

```

```python
pol.to_csv('Policy_frequency.csv')
```
