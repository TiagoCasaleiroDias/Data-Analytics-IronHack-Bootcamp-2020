```python
def create_ranks(df,job):
        df=df[(df.job)==job]
        df.drop(columns=['job'],inplace=True)
        df=df.sort_values(by='offers',ascending=False).reset_index(drop=True)
        df['rank']=df['offers'].rank(method='first',ascending=False)
        return df

data_scientist_rank=create_ranks(final,'Data Scientist')
data_analyst_rank=create_ranks(final,'Data Analyst')
business_analyst_rank=create_ranks(final,'Business Analyst')
data_engineer_rank=create_ranks(final,'Data Engineer')

data_analyst_rank
```
